```python
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu
def scorer(result,answer):
    
    result = result.lower()
    answer = answer.lower()
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2'], use_stemmer=True)
    scores = scorer.score(result,answer)
    rouge1_precision = scores['rouge1'][0]
    rouge1_recall =scores['rouge1'][1]
    rouge1_fmeasure = scores['rouge1'][2]
    rouge2_precision = scores['rouge2'][0]
    rouge2_recall =scores['rouge2'][1]
    rouge2_fmeasure=scores['rouge2'][2]

    answer_token = [nltk.word_tokenize(answer) for sent in [answer]]
    result_token = [nltk.word_tokenize(result)]
    
    bleu_score = sentence_bleu(result, answer)
    
    return rouge1_precision,rouge1_recall,rouge1_fmeasure,rouge2_precision,rouge2_recall,rouge2_fmeasure,bleu_score
```
