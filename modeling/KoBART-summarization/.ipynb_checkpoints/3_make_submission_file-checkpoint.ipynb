# submission file 

```python
import pickle
import pandas as pd

with open("/home/ubuntu/project/tmp/KoBART-summarization/answer_l.pkl","rb") as fr:
    data1 = pickle.load(fr)

with open("/home/ubuntu/project/tmp/KoBART-summarization/answer_s1.pkl","rb") as fr:
    data2 = pickle.load(fr)

with open("/home/ubuntu/project/tmp/KoBART-summarization/answer_s2.pkl","rb") as fr:
    data3 = pickle.load(fr)
```

## dictionary들을 종합한 dic을 생성합니다.

```python
dic = {}
dic.update(data1)
dic.update(data2)
dic.update(data3)

len(dic)
```

```python
test_data = pd.read_csv('KoBART-summarization/data/test.csv')
test_data
```

## test 데이터의 id를 기준으로 모델의 summary 결과를 추가해 줍니다.

```python
summary = []

ids = list(test_data['id'])

for id in ids:
  summary.append(dic[id])

test_data['summary'] = summary
```

```python
test_data
```

## 대회 결과 제출 양식에 맞게 파일을 저장해 줍니다.

```python
test_data = test_data[['id', 'summary']]
test_data
```

```python
test_data.to_csv('/home/ubuntu/project/tmp/KoBART-summarization/answer.csv', index=False, encoding='utf-8-sig')
```
